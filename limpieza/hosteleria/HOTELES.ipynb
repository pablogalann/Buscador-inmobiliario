```python
def limpiar_tipo(x):
    if 'RESTAURANTE'in x:
        return 'HOTEL'
    if x == 'ALBERGUES JUVENILES':
        return 'ALBERGUE JUVENIL'
    if x == 'APART-HOTELES':
        return 'APART-HOTEL'
    if x == 'VIVIENDAS TURISTICAS':
        return 'VIVIENDA TURISTICA'
    if x == 'COLEGIOS MAYORES Y RESIDENCIAS DE ESTUDIANTES':
        return 'COLEGIO MAYOR/RESIDENCIA DE ESTUDIANTE'
    if x == 'HOSTALES':
        return 'HOSTAL'
    if x == 'CASAS DE HUESPEDES':
        return 'CASA DE HUESPEDES'
    if x == 'PENSIONES':
        return 'PENSION'
    if x == 'CAMPINGS Y APARCAMIENTOS PARA CARAVANAS':
        return 'CAMPING Y APARCAMIENTO PARA CARAVANAS'
    else:
        return x
```

```python
hot.tipo_alojamiento=hot.tipo_alojamiento.apply(limpiar_tipo)
```
