```python
def limpiar_acento(x):
    
    if 'Ã\x93' in x:
        x = x.replace('Ã\x93','O')
    if 'Ã\x8d' in x:
        x = x.replace('Ã\x8d','I')
    if 'Ã\x89' in x:
        x = x.replace('Ã\x89','E')
    if 'Ã\x81' in x:
        x = x.replace('Ã\x81','A')
    if 'Ã\x9a' in x:
        x = x.replace('Ã\x9a','U')
   
        
    return x
        
```
