```python
def limpiar(x):
    if 'Ã' in x:
        x= x.replace('Ã','i')
        return x
    if 'Ã¡' in x:
        x= x.replace('Ã¡','a')
        return x
    if 'Ã©' in x:
        x= x.replace('Ã©','e')
        return x
    if 'Ãº' in x:
        x= x.replace('Ãº','u')
        return x
    if 'Ã³' in x:
        x= x.replace('Ã³','u')
        return x
    if 'i±' in x:
        x= x.replace('i±','n')
        return x
    else:
        return x
    
```

```python
oc.nombre_equip = oc.nombre_equip.apply(limpiar)
```
