```python
aud_lista = []

for x in aud.nombre_equip:
    
    if 'Sala' in x:
        aud_lista.append('SALA CONCIERTOS')
    elif 'Café' in x:
        aud_lista.append('SALA CONCIERTOS')
    elif 'Teatro' in x:
        aud_lista.append('TEATRO')
    elif 'Club' in x:
        aud_lista.append('SALA CONCIERTOS')
    elif 'Auditorio' in x:
        aud_lista.append('AUDITORIO')
    elif 'Tablao' in x:
        aud_lista.append('TABLAO')
    else:
        aud_lista.append('USO POLIVALENTE')

print(aud_lista)
```
